## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
two["Gender"] = label.fit_transform(two["Gender"])
two["Married"] = label.fit_transform(two["Married"])
two["Education"] = label.fit_transform(two["Education"])
two["Self_Employed"] = label.fit_transform(two["Self_Employed"])
two["Property_Area"] = label.fit_transform(two["Property_Area"])
two["Loan_Status"] = label.fit_transform(two["Loan_Status"])
```

## Correlation 

```python
corr_matrix = two.corr()
corr_matrix['Loan_Status'].sort_values(ascending = False)
```

```python
import numpy as np
loan_num = two.drop("Loan_Status", axis = 1)
loan_labels = two["Loan_Status"].copy()
loan_num_arr = loan_num.to_numpy()
loan_labels_arr = loan_labels.to_numpy()
```

## Introducing Xgboost Model for predicting Whether Loan should be given to the person or not

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(random_state = 5, learning_rate = 1, max_depth = 2)
```

```python
model.fit(loan_num_arr, loan_labels_arr)
```

```python
y_predict = model.predict(loan_num_arr)
y_predict
```

```python
list(loan_labels_arr)
```

## Accuracy of the model

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(loan_labels_arr, y_predict))
```

```python
pd.crosstab(loan_labels_arr, y_predict)
```

## Creating Joblib File of the model for Showing Final Predictions
